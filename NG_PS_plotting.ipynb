# Plotting and Visualization in Python 

##### Matplotlib is a versatile Python library that generates plots for data visualization

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

## Simple plot

```python
x=np.arange(1,11)
```

```python
x
```

```python
plt.plot(x)
plt.show()
```

```python
#create a new array with squared values of x
y=x**2
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,label='X^2')
plt.legend()
plt.show()
plt.savefig('x2.png')
```

## Save the plot as 'png' file

```python
plt.savefig('x_square.png')
```

## Figures 

A basic Matplotlib figure is made up of multiple components common to different plot types. Plots in matplotlib reside within a Figure object. You can create a new figure with plt.figure()

A figure is the whole plotting area that contains all plot elements. Multiple subplots may be tiled in grid within one figure

```python
fig=plt.figure()
```

#### add plot to the figure object

```python
ax1 = fig.add_subplot(2, 2,1)
```

####### This means that the figure should be 2 × 2 (so up to four plots in total), and we’re selecting the first of four subplots (numbered from 1).

```python
ax2 = fig.add_subplot(2, 2,2)
ax3 = fig.add_subplot(2, 2,3)
plt.plot()
plt.show()
```
