## Arguments (*args) & Keyword Arguments(**kwars)

```python
#create simple function with positional arguments 

def five_percent(a,b):
    
    # a,b are called positional arguments 
    """
    
    DOCSTRING : This function returns 5 % value of total sum of input
    
    INPUT : Integer values
    
    OUTPUT :  5 % of total input values
    
    """
    val=sum((a,b))*0.05
    
    return val
```

```python
five_percent(1000,750)
```

1000 is assigned to 'a' because it was at the first positon and 750 assigned to 'b' becuase it was at second position in function

* Incase where we are not sure of the number of arguments/parameters in advanvce , we use *args & **kwargs
* It allows us to take arbitary number of arguments with out declaring in advance


```python
#create simple function with arbitary arguments 

def five_percent(*args):
    
    
    """
    
    DOCSTRING : This function returns 5 % value of total sum of input
    
    INPUT : Integer values
    
    OUTPUT :  5 % of total input values
    
    """
    val=sum((args))*0.05
    
    return val
```

```python
#use the function by passing any number of arguments
five_percent(300,500,100,400,1000,4000,6000,66)
```

```python
#### one case use any keyword insteasd of args as well , but it is used widely by most of the developer community

#create simple function with arbitary arguments 

def five_percent(*london):
    
    
    """
    
    DOCSTRING : This function returns 5 % value of total sum of input
    
    INPUT : Integer values
    
    OUTPUT :  5 % of total input values
    
    """
    val=sum((london))*0.05
    
    return val
```

```python
five_percent(33,55,66,77,88)
```

```python
#In the case when the input parameters are key value pairs /dictionary

def capital(*args,**kwargs):
    
    """
    
    DOCSTRING : This function returns capital of the country
    
    INPUT : Key value pairs of Country and Capital 
    
    OUTPUT :  Capital of the country
    
    """
    
    
    
    print("The capital of {} is {}".format(args[1],kwargs['UK']))
```

```python
capital('India','UK','France',India='Delhi',UK='London',France='Paris')
```

```python
try:
    f=open('testfile','w')
    f.write('This is a test line')
except TypeError:
    print( "Hey there is a Type error ")
except OSError:
    print("Hey you have an OS error")
#Runs irrespective of error
finally:
    print ("Able to execute")

```

```python
try:
    #change the permission to read only
    f=open('testfile','r')
    f.write('This is a test line')
except TypeError:
    print( "Hey there is a Type error ")
except OSError:
    print("Hey you have an OS error")
#Runs irrespective of error
finally:
    print ("Able to execute")
```

#### Try/except/break/finally inside of a function 

```python
def ask_for_age():
    
    try:
        result=int(input("Please provide your age "))
        
    except:
        print("Oops ...That doesn't seems to be an integer ")
        
    finally:
        print("End of this session ")
```

```python
ask_for_age()
```

```python
ask_for_age()
```

```python
def ask_for_age():
    while True:
        try:
            result=int(input("Please provide your age "))
        except:
            print("Oops ...That doesn't seems to be an integer ")
            continue
        else:
            print("Thank you...that works ")
            print("Im, done")
            break
        finally:
            print("End of this session ")
```
